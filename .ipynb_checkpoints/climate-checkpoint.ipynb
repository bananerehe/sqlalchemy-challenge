```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
base = automap_base()

# reflect the tables
base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
base.classes.keys()
```

```python
# Save references to each table
measurement = base.classes.measurement
station = base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent_date = session.query(measurement.date).order_by(measurement.date.desc()).first()
most_recent_date = str(most_recent_date)[2:-3]
print(most_recent_date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
one_year_previous = str(eval(most_recent_date[0:4])-1) + most_recent_date[4:]
print(one_year_previous)
```

```python
# Perform a query to retrieve the data and precipitation scores
previous_year_precip = session.query(measurement.date, measurement.prcp).\
    filter(measurement.date >= one_year_previous).filter(measurement.date <= most_recent_date).order_by(measurement.date).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
previous_year_df = pd.DataFrame(previous_year_precip, columns=["date", "prcp"])
previous_year_df.set_index("date", inplace=True)
previous_year_df.rename(columns={"prcp":"precipitation"}, inplace=True)
previous_year_df.head()
```

```python
# Sort the dataframe by date
previous_year_df.sort_values("date")

# Use Pandas Plotting with Matplotlib to plot the data
previous_year_df.plot()
plt.rcParams["figure.figsize"] = (20, 6)
plt.legend(loc=9)
plt.savefig("precipitation_analysis.png")
plt.xlabel("Date", fontsize=16, labelpad=20)
plt.ylabel("Precipitation (in)", fontsize=16, labelpad=20)
plt.title("Maximum Precipitation for One Year \n Honolulu, Hawaii")
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
previous_year_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
total_stations = session.query(station).count()
print(total_stations)
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
observations = session.query(measurement.station, func.count(measurement.tobs))\
             .group_by(measurement.station).order_by(func.count(measurement.tobs).desc())

for station in observations:
    print(station)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
most_active_station = observations[0][0]

stats = session.query(func.min(measurement.tobs), func.max(measurement.tobs),
                      func.avg(measurement.tobs)).filter(measurement.station == most_active_station)
                      
stats_list = list(stats)

print(stats_list)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
tobs = session.query(measurement.tobs)\
.filter(measurement.date >= one_year_previous)\
.filter(measurement.date <= most_recent_date)\
.filter(measurement.station == most_active_station)\

tobs_list = []
for tob in tobs:
    tobs_list.append(tob[0])
    
fig = plt.figure()
plt.hist(tobs_list, bins=12, label="TOBS")
plt.legend(loc=1)
plt.rcParams["figure.figsize"] = (20, 6)
plt.xlabel("Temperature (F)", fontsize=16, labelpad=20)
plt.ylabel("Frequency", fontsize=16, labelpad=20)
plt.title("12-Month Temperature Observation \n Honolulu, Hawaii")
plt.savefig("station_observation_analysis.png")
plt.show()
```
